```python
class Proofer:
    def __init__(self, name):
        self.name = name
        self.proofs = []

    def submit_proof(self, proof):
        self.proofs.append(proof)

    def get_proofs(self):
        return self.proofs


class Market:
    def __init__(self):
        self.proofers = []

    def add_proofer(self, proofer):
        self.proofers.append(proofer)

    def get_all_proofs(self):
        all_proofs = []
        for proofer in self.proofers:
            all_proofs.extend(proofer.get_proofs())
        return all_proofs


# Usage example
market = Market()

proofer1 = Proofer("Proofer 1")
proofer1.submit_proof("Proof 1")
proofer1.submit_proof("Proof 2")

proofer2 = Proofer("Proofer 2")
proofer2.submit_proof("Proof 3")

market.add_proofer(proofer1)
market.add_proofer(proofer2)

all_proofs = market.get_all_proofs()
print(all_proofs)
```
